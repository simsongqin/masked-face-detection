```python
# Image Loading Code used for these examples
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
import time
```

```python
dir ='/Users/simsongqin/Downloads/CASIA-maxpy-clean/'
output_dir = '/Users/simsongqin/Downloads/CASIA-annotated-sin/'
image_lists = os.listdir(dir)
#print(image_lists)
num_of_image_annotated = 0

f= open("CASIA-annotated-sin"+str(time.time())+".txt","w+")
for j in image_lists:
    path = os.path.join(output_dir, j)
    if(os.path.exists(path)):
        continue
    os.mkdir(path)
    temp = '/Users/simsongqin/Downloads/CASIA-maxpy-clean/'+j
    #print(j)
    for i in os.listdir(temp):
        #print(i)
        img = cv2.imread('/Users/simsongqin/Downloads/CASIA-maxpy-clean/'+j+'/'+i, cv2.IMREAD_UNCHANGED)
        #print(len(img.shape))
        
        num_of_image_annotated+=1
        if(num_of_image_annotated%1000==0):
            print("Number of image annotated: ",num_of_image_annotated)
        if(img.shape==None):
            print(i)
            continue
        WIDTH = img.shape[0]
        HEIGHT = img.shape[1]
        if (len(img.shape)==2):
            
            for a in range(74,112):
                for b in range(32,64):
                        if (img[a][b] != 0):
                            img[a][b] = 0

            cv2.imwrite('/Users/simsongqin/Downloads/CASIA-annotated-sin/'+j+'/'+i,img) 
            f.write('/'+j+'/'+i+'\n')
            continue

        DEPTH = img.shape[2]
        
        for a in range(74,112):
            for b in range(32,64):
                for c in range(DEPTH):
                    if (img[a][b][c] != 0):
                        img[a][b][c] = 0
        
        cv2.imwrite('/Users/simsongqin/Downloads/CASIA-annotated-sin/'+j+'/'+i,img)
        f.write('/'+j+'/'+i+'\n')

print(num_of_image_annotated)
f.close()
```
